```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import math
from PIL import Image
import warnings
warnings.filterwarnings('ignore')
```

```python
directory = 'trainingSet'
image_list = []
labels = []
list_folders = os.listdir(directory)
for folders in list_folders:
    images = os.listdir(f'{directory}/{folders}')
    for image in images:
        image_dir = f'{directory}/{folders}/{image}'
        image_list.append(image_dir)
        labels.append(folders)
        

#randomizing the dataset
image_list = np.array(image_list)
labels = np.array(labels)
randomize = np.arange(len(labels))
np.random.shuffle(randomize)
image_list = image_list[randomize]
labels = labels[randomize]

#splitting the dataset into training and testing part
train_image_list = image_list[0:int(len(image_list)*0.9)]
test_image_list = image_list[int(len(image_list)*0.9):]
train_labels = labels[0:int(len(image_list)*0.9)]
test_labels = labels[int(len(image_list)*0.9):]
```

```python
def preprocess_input(img):
    img = np.array(img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
    y, u, v = cv2.split(img)
    y = y / 255.0
    x = np.array(y.flatten()).reshape(1,784)
    return x
```

```python
#Activation Functions
def sigmoid(x):
    return 1. / (1. + np.exp(-x))

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=1)

def tanh(x):
    return math.tanh(x)

def relu(x):
    return x * (x > 0)
```

# Neural Network with a Single Hidden Layer

Input Size = 784
First Layer Input = 784
First Layer Output = 32
Second Layer Input = 32
Second Layer Output = 10

### Our Architecture


```python
input_layer = 784
first_layer_out = 32
second_layer_out = 10

def initialize_weights_and_bias(mode):
    if mode == 'random':
        W1 = np.random.random((first_layer_out, input_layer)).astype(np.float32)
        b1 = np.random.random((1, first_layer_out)).astype(np.float32)
        W2 = np.random.random((second_layer_out, first_layer_out)).astype(np.float32)
        b2 = np.random.random((1, second_layer_out)).astype(np.float32)
    elif mode == 'xavier':
        W1 = np.random.normal(loc=0.0, scale=np.sqrt(2/(first_layer_out+input_layer)), size=(first_layer_out, input_layer)).astype(np.float32)
        b1 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1+first_layer_out)), size=(1, first_layer_out)).astype(np.float32)
        W2 = np.random.normal(loc=0.0, scale=np.sqrt(2/(second_layer_out+first_layer_out)), size=(second_layer_out, first_layer_out)).astype(np.float32)
        b2 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1+second_layer_out)), size=(1, second_layer_out)).astype(np.float32)
    elif mode == 'uniform':
        W1 = np.random.uniform(low = -np.sqrt(6/(first_layer_out+input_layer)), high = np.sqrt(6/(first_layer_out+second_layer_out)), size=(first_layer_out, input_layer)).astype(np.float32)
        b1 = np.random.uniform(low = -np.sqrt(6/(1+first_layer_out)), high = np.sqrt(6/(second_layer_out+second_layer_out)), size=(1, first_layer_out)).astype(np.float32)
        W2 = np.random.uniform(low = -np.sqrt(6/(second_layer_out+first_layer_out)), high = np.sqrt(6/(second_layer_out+second_layer_out)), size=(second_layer_out, first_layer_out)).astype(np.float32)
        b2 = np.random.uniform(low = -np.sqrt(6/(1+second_layer_out)), high = np.sqrt(6/(second_layer_out+second_layer_out)), size=(1, second_layer_out)).astype(np.float32)
    elif mode == 'constant':
        W1 = np.ones((first_layer_out, input_layer))
        b1 = np.ones((1, first_layer_out))
        W2 = np.ones((second_layer_out, first_layer_out))
        b2 = np.ones((1, second_layer_out))
    return W1, b1, W2, b2
```

```python
def forward_pass(x, W1, b1, W2, b2):
    
    #First layer
    x = np.array(x, dtype=np.float32)
    res1 = np.matmul(x, W1.T) + b1
    layer1_out = sigmoid(res1)
    # print(layer1_out)
    
    #Second layer
    res2 = np.matmul(layer1_out, W2.T) + b2
    layer2_out = softmax(res2)
    #print(layer2_out)
    return layer2_out.flatten(), layer1_out.flatten()
```

```python
def backward_pass(x, W1, b1, W2, b2, y_hat, y, z1, lr = 1):
    
    x = np.array(x, dtype= np.float32).reshape(1,784)
    y = np.array(y, dtype = np.float32).reshape(1,10)
    y_hat = np.array(y_hat, dtype=np.float32).reshape(1,10)
    z1 = np.array(z1, dtype=np.float32).reshape(1,32)
    
    db2 = (y_hat-y)                                                    #(1x10).(10x1) = (1x1).(1x10) = (1x10)
    dw2 = np.matmul(db2.T, z1)                                         #(10x1)(1x32)  = 10x32
    db1 = np.matmul(np.matmul(np.matmul(db2, W2).T, z1), (1-z1).T)     #(1x10).(10x32)=(32x1).(1x32)=(32x32).(32x1)=(32x1)
    dw1 = np.matmul(db1, x)                                            #(32x1).(1x784) = 32x784
    
    up_W2 = W2 - lr*dw2
    up_b2 = b2 - lr*db2
    up_b1 = b1 - lr*db1.T
    up_W1 = W1 - lr*dw1
    
    return up_W2, up_b2, up_W1, up_b1
    
```

```python
# print("Predicted Class | Actual Class")
import matplotlib.pyplot as plt


W1, b1, W2, b2 = initialize_weights_and_bias('uniform')
target = np.array(train_labels)

cost = 0
losses = []
epochs = 10

for j in range(epochs):
    losss = 0
    tp = 0
    for i in range(len(train_image_list)):
        
        img = cv2.imread(train_image_list[i])
        x = preprocess_input(img)
        y = np.zeros(10)
        y[int(target[i])] = 1
        
    

        #network forward pass
        y_hat, z1 = forward_pass(x, W1, b1, W2, b2)
        
        
        #we use categorical cross entropy loss
        loss = 0
        for i in range(len(y)):
            if y[i] != 0:
                loss +=  -y[i] * np.log(y_hat[i])
        losss += loss
        
        
        #the true positives for this case, to calculate accuracy
        if np.argmax(y_hat) == np.argmax(y):
            tp += 1
        
        #network backward pass
        W2, b2, W1, b1 = backward_pass(x, W1, b1, W2, b2, y_hat, y, z1)
    
    #we append total loss of each epoch to the losses array
    losses.append(losss)
    print(f"Epoch = {j}, Accuracy = {tp/len(train_image_list)}, loss= {losss}")
```

```python
# print(losses)
plt.plot(losses)
plt.show()
```

```python
print(f"Accuracy for training data: {round(tp/len(train_image_list), 2)*100}%")
```

# Neural Network with 2 Hidden Layers


```python
input_layer = 784
first_layer_out = 64
second_layer_out = 32
third_layer_out = 10

def initialize_weights_and_bias(mode):
    if mode == 'random':
        W1 = np.random.random((first_layer_out, input_layer)).astype(np.float32)
        b1 = np.random.random((1, first_layer_out)).astype(np.float32)
        W2 = np.random.random((second_layer_out, first_layer_out)).astype(np.float32)
        b2 = np.random.random((1, second_layer_out)).astype(np.float32)
        W3 = np.random.random((third_layer_out, second_layer_out)).astype(np.float32)
        b3 = np.random.random((1, third_layer_out)).astype(np.float32)
    elif mode == 'xavier':
        W1 = np.random.normal(loc=0.0, scale=np.sqrt(2/(first_layer_out+input_layer)), size=(first_layer_out, input_layer)).astype(np.float32)
        b1 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1+first_layer_out)), size=(1, first_layer_out)).astype(np.float32)
        W2 = np.random.normal(loc=0.0, scale=np.sqrt(2/(second_layer_out+first_layer_out)), size=(second_layer_out, first_layer_out)).astype(np.float32)
        b2 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1+second_layer_out)), size=(1, second_layer_out)).astype(np.float32)
        W3 = np.random.normal(loc=0.0, scale=np.sqrt(2/(third_layer_out+second_layer_out)), size=(third_layer_out, second_layer_out)).astype(np.float32)
        b3 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1+third_layer_out)), size=(1, third_layer_out)).astype(np.float32)
    elif mode == 'uniform':
        W1 = np.random.uniform(low = -np.sqrt(6/(first_layer_out+input_layer)), high = np.sqrt(6/(first_layer_out+second_layer_out)), size=(first_layer_out, input_layer)).astype(np.float32)
        b1 = np.random.uniform(low = -np.sqrt(6/(1+first_layer_out)), high = np.sqrt(6/(second_layer_out+second_layer_out)), size=(1, first_layer_out)).astype(np.float32)
        W2 = np.random.uniform(low = -np.sqrt(6/(second_layer_out+first_layer_out)), high = np.sqrt(6/(second_layer_out+second_layer_out)), size=(second_layer_out, first_layer_out)).astype(np.float32)
        b2 = np.random.uniform(low = -np.sqrt(6/(1+second_layer_out)), high = np.sqrt(6/(second_layer_out+second_layer_out)), size=(1, second_layer_out)).astype(np.float32)
        W3 = np.random.uniform(low = -np.sqrt(6/(third_layer_out+second_layer_out)), high = np.sqrt(6/(third_layer_out+third_layer_out)), size=(third_layer_out, second_layer_out)).astype(np.float32)
        b3 = np.random.uniform(low = -np.sqrt(6/(1+third_layer_out)), high = np.sqrt(6/(third_layer_out+third_layer_out)), size=(1, third_layer_out)).astype(np.float32)
    elif mode == 'constant':
        W1 = np.ones((first_layer_out, input_layer))
        b1 = np.ones((1, first_layer_out))
        W2 = np.ones((second_layer_out, first_layer_out))
        b2 = np.ones((1, second_layer_out))
        W3 = np.ones((third_layer_out, second_layer_out))
        b3 = np.ones((1, third_layer_out))
    return W1, b1, W2, b2, W3, b3
```

```python
def forward_pass(x, W1, b1, W2, b2, W3, b3):
    
    #First layer
    x = np.array(x, dtype=np.float32)
    res1 = np.matmul(x, W1.T) + b1
    layer1_out = sigmoid(res1)
    # print(layer1_out)
    
    #Second layer
    res2 = np.matmul(layer1_out, W2.T) + b2
    layer2_out = sigmoid(res2)
    #print(layer2_out)

    #Third Layer
    res3 = np.matmul(layer2_out, W3.T) + b3
    layer3_out = softmax(res3)
    #print(layer2_out)
    
    return layer3_out.flatten(), layer2_out.flatten(), layer1_out.flatten()
```

```python
def backward_pass(x, W1, b1, W2, b2, W3, b3, y_hat, y, z1, z2, lr = 0.01):
    
    x = np.array(x, dtype= np.float32).reshape(1,784)
    y = np.array(y, dtype = np.float32).reshape(1,10)
    y_hat = np.array(y_hat, dtype=np.float32).reshape(1,10)
    z1 = np.array(z1, dtype=np.float32).reshape(1,64)
    z2 = np.array(z2, dtype=np.float32).reshape(1, 32)
    
    db3 = (y_hat-y)                                                    #(1x10)-(1x10) = (1x10)
    dw3 = np.matmul(db3.T, z2)                                         #(10x1)(1x32)  = 10x32
    db2 = np.matmul(np.matmul(np.matmul(db3, W3).T, z2), (1-z2).T).T   #(1x10).(10x32)=(32x1).(1x32)=(32x32).(32x1)=(1x32)
    dw2 = np.matmul(db2.T, z1)                                         #(32x1).(1x64) = 32x64
    db1 = np.matmul(np.matmul(np.matmul(db2, W2).T, z1), (1-z1).T).T   #(1x32).(32x64)=(64x1)(1x64)=(64x64).(64x1)=(1x64)
    dw1 = np.matmul(db1.T, x)                                          #(64x1).(1x784)=(64x784) 

    
    up_W3 = W3 - lr*dw3
    up_b3 = b3 - lr*db3
    up_W2 = W2 - lr*dw2
    up_b2 = b2 - lr*db2
    up_b1 = b1 - lr*db1
    up_W1 = W1 - lr*dw1
    
    return up_W3, up_b3, up_W2, up_b2, up_W1, up_b1
```

```python
# print("Predicted Class | Actual Class")
import matplotlib.pyplot as plt


W1, b1, W2, b2, W3, b3 = initialize_weights_and_bias('uniform')
target = np.array(train_labels)

cost = 0
losses = []
epochs = 10

for j in range(epochs):
    losss = 0
    tp = 0
    for i in range(len(train_image_list)):
        
        img = cv2.imread(train_image_list[i])
        x = preprocess_input(img)
        y = np.zeros(10)
        y[int(target[i])] = 1
        
    

        #network forward pass
        y_hat, z2, z1 = forward_pass(x, W1, b1, W2, b2, W3, b3)
        
        
        #we use categorical cross entropy loss
        loss = 0
        for i in range(len(y)):
            if y[i] != 0:
                loss +=  -y[i] * np.log(y_hat[i])
        losss += loss
        
        
        #the true positives for this case, to calculate accuracy
        if np.argmax(y_hat) == np.argmax(y):
            tp += 1
        
        #network backward pass
        W3, b3, W2, b2, W1, b1 = backward_pass(x, W1, b1, W2, b2, W3, b3, y_hat, y, z1, z2)
    
    #we append total loss of each epoch to the losses array
    losses.append(losss)
    print(f"Epoch = {j}, Accuracy = {tp/len(train_image_list)}, loss= {losss}")
```

```python
# print(losses)
plt.plot(losses)
plt.show()
```

 # Network with 5 Hidden Layers

```python
input_layer = 784
hidden_layer1 = 392
hidden_layer2 = 196
hidden_layer3 = 64
hidden_layer4 = 32
output_layer = 10

def initialize_weights_and_bias(mode, distribution='uniform'):
    '''
    Initialize weights and bias for neural network layers using the specified distribution. 
    '''

    if mode == 'random':
        W1 = np.random.random((hidden_layer1, input_layer)).astype(np.float32)   #784x392
        b1 = np.random.random((1, hidden_layer1)).astype(np.float32)             #1x392
        W2 = np.random.random((hidden_layer1, hidden_layer2)).astype(np.float32) #392x196
        b2 = np.random.random((1, hidden_layer2)).astype(np.float32)             #1x196
        W3 = np.random.random((hidden_layer2, hidden_layer3)).astype(np.float32) #196x64
        b3 = np.random.random((1, hidden_layer3)).astype(np.float32)             #1x64
        W4 = np.random.random((hidden_layer3, hidden_layer4)).astype(np.float32) #64x32
        b4 = np.random.random((1, hidden_layer4)).astype(np.float32)             #1x32
        W5 = np.random.random((hidden_layer4, output_layer)).astype(np.float32)  #32x10
        b5 = np.random.random((1, output_layer)).astype(np.float32)              #1x10
    elif mode == 'xavier':
        if distribution == 'normal':
            W1 = np.random.normal(loc=0.0, scale=np.sqrt(2/(hidden_layer1 + input_layer)), size=((hidden_layer1, input_layer))).astype(np.float32)    #784x392
            b1 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1+hidden_layer1)), size=((1, hidden_layer1))).astype(np.float32)              #1x392   
            W2 = np.random.normal(loc=0.0, scale=np.sqrt(2/(hidden_layer1 + hidden_layer2)), size=((hidden_layer1, hidden_layer2))).astype(np.float32)  #392x196
            b2 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1 + hidden_layer2)), size=((1, hidden_layer2))).astype(np.float32)              #1x196
            W3 = np.random.normal(loc=0.0, scale=np.sqrt(2/(hidden_layer2 + hidden_layer3)), size=((hidden_layer2, hidden_layer3))).astype(np.float32)  #196x64
            b3 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1 + hidden_layer3)), size=((1, hidden_layer3))).astype(np.float32)              #1x64
            W4 = np.random.normal(loc=0.0, scale=np.sqrt(2/(hidden_layer3 + hidden_layer4)), size=((hidden_layer3, hidden_layer4))).astype(np.float32)  #64x32
            b4 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1 + hidden_layer4)), size=((1, hidden_layer4))).astype(np.float32)              #1x32
            W5 = np.random.normal(loc=0.0, scale=np.sqrt(2/(hidden_layer4 + output_layer)), size=((hidden_layer4, output_layer))).astype(np.float32)   #32x10
            b5 = np.random.normal(loc=0.0, scale=np.sqrt(2/(1 + output_layer)), size=((1, output_layer))).astype(np.float32)               #1x10
        elif distribution == 'uniform':
            W1 = np.random.uniform(low = -np.sqrt(6/(hidden_layer1 + input_layer)), high = np.sqrt(6/(hidden_layer1 + input_layer)), size=((hidden_layer1, input_layer))).astype(np.float32)   #392x784
            b1 = np.random.uniform(low = -np.sqrt(6/(1+hidden_layer1)), high = np.sqrt(6/(1+hidden_layer1)), size=((1, hidden_layer1))).astype(np.float32)             #1x392
            W2 = np.random.uniform(low = -np.sqrt(6/(hidden_layer1 + hidden_layer2)), high = np.sqrt(6/(hidden_layer1 + hidden_layer2)), size=((hidden_layer2, hidden_layer1))).astype(np.float32) #392x196
            b2 = np.random.uniform(low = -np.sqrt(6/(1 + hidden_layer2)), high = np.sqrt(6/(1 + hidden_layer2)), size=((1, hidden_layer2))).astype(np.float32)             #1x196
            W3 = np.random.uniform(low = -np.sqrt(6/(hidden_layer2 + hidden_layer3)), high = np.sqrt(6/(hidden_layer2 + hidden_layer3)), size=((hidden_layer3, hidden_layer2))).astype(np.float32) #196x64
            b3 = np.random.uniform(low = -np.sqrt(6/(1 + hidden_layer3)), high = np.sqrt(6/(1 + hidden_layer3)), size=((1, hidden_layer3))).astype(np.float32)             #1x64
            W4 = np.random.uniform(low = -np.sqrt(6/(hidden_layer3 + hidden_layer4)), high = np.sqrt(6/(hidden_layer3 + hidden_layer4)), size=((hidden_layer4, hidden_layer3))).astype(np.float32) #64x32
            b4 = np.random.uniform(low = -np.sqrt(6/(1 + hidden_layer4)), high = np.sqrt(6/(1 + hidden_layer4)), size=((1, hidden_layer4))).astype(np.float32)             #1x32
            W5 = np.random.uniform(low = -np.sqrt(6/(hidden_layer4 + output_layer)), high = np.sqrt(6/(hidden_layer4 + output_layer)), size=((output_layer, hidden_layer4))).astype(np.float32)  #32x10
            b5 = np.random.uniform(low = -np.sqrt(6/(1 + output_layer)), high = np.sqrt(6/(1 + output_layer)), size=((1, output_layer))).astype(np.float32)              #1x10
    elif mode == 'constant':
        W1 = np.ones((hidden_layer1, input_layer)).astype(np.float32)     #784x392
        b1 = np.ones((1, hidden_layer1)).astype(np.float32)               #1x392
        W2 = np.ones((hidden_layer1, hidden_layer2)).astype(np.float32)   #392x196
        b2 = np.ones((1, hidden_layer2)).astype(np.float32)               #1x196
        W3 = np.ones((hidden_layer2, hidden_layer3)).astype(np.float32)   #196x64
        b3 = np.ones((1, hidden_layer3)).astype(np.float32)               #1x64
        W4 = np.ones((hidden_layer3, hidden_layer4)).astype(np.float32)   #64x32
        b4 = np.ones((1, hidden_layer4)).astype(np.float32)               #1x32 
        W5 = np.ones((hidden_layer4, output_layer)).astype(np.float32)    #32x10
        b5 = np.ones((1, output_layer)).astype(np.float32)                #1x10
    elif mode == 'zero':
        W1 = np.zeros((hidden_layer1, input_layer)).astype(np.float32)
        b1 = np.zeros((1, hidden_layer1)).astype(np.float32)
        W2 = np.zeros((hidden_layer1, hidden_layer2)).astype(np.float32)
        b2 = np.zeros((1, hidden_layer2)).astype(np.float32)
        W3 = np.zeros((hidden_layer2, hidden_layer3)).astype(np.float32)
        b3 = np.zeros((1, hidden_layer3)).astype(np.float32)
        W4 = np.zeros((hidden_layer3, hidden_layer4)).astype(np.float32)
        b4 = np.zeros((1, hidden_layer4)).astype(np.float32)
        W5 = np.zeros((hidden_layer4, output_layer)).astype(np.float32)
        b5 = np.zeros((1, output_layer)).astype(np.float32)
    else:
        print('Error: mode not recognized')
        return

    return W1, b1, W2, b2, W3, b3, W4, b4, W5, b5  
```

```python
def forward_pass(x, W1, b1, W2, b2, W3, b3, W4, b4, W5, b5):
    
    #First layer
    x = np.array(x, dtype=np.float32)                 #784x1
    
    res1 = np.matmul(x, W1.T) + b1                    #392x1
    layer1_out = sigmoid(res1)                        #392x1
    # print(layer1_out)                 
    
    #Second layer
    res2 = np.matmul(layer1_out, W2.T) + b2          #196x1
    layer2_out = sigmoid(res2)                       #196x1
    #print(layer2_out)                                
    
    #Third layer
    res3 = np.matmul(layer2_out, W3.T) + b3         #64x1
    layer3_out = sigmoid(res3)                      #64x1
    #print(layer3_out)
    
    #Fourth Layer
    res4 = np.matmul(layer3_out, W4.T) + b4        #32x1
    layer4_out = sigmoid(res4)                     #32x1
    #print(layer4_out)
    
    #Fifth Layer
    res5 = np.matmul(layer4_out, W5.T) + b5        #10x1
    layer5_out = softmax(res5)                     #10x1
    #print(layer5_out)
    
    
    return layer1_out.flatten(), layer2_out.flatten(), layer3_out.flatten(), layer4_out.flatten(), layer5_out.flatten()
```

```python
def backward_pass(x, y, W1, b1, W2, b2, W3, b3, W4, b4, W5, b5, z1, z2, z3, z4, y_hat, lr=0.0001):
    
    x = np.array(x, dtype= np.float32).reshape(1,784)
    y = np.array(y, dtype = np.float32).reshape(1,10)
    y_hat = np.array(y_hat, dtype=np.float32).reshape(1,10)
    z1 = np.array(z1, dtype=np.float32).reshape(1, 392)
    z2 = np.array(z2, dtype=np.float32).reshape(1, 196)
    z3 = np.array(z3, dtype=np.float32).reshape(1, 64)
    z4 = np.array(z4, dtype=np.float32).reshape(1, 32)

    
    #Final Layer
    db5 = (y_hat-y)                                              
    dw5 = np.matmul(db5.T, z4)                                         
    
    #Fourth Layer
    db4 = np.matmul(np.matmul(np.matmul(db5, W5).T, z4), (1-z4).T).T   
    dw4 = np.matmul(db4.T, z3)                                         
    
    #Third Layer
    db3 = np.matmul(np.matmul(np.matmul(db4, W4).T, z3), (1-z3).T).T   
    dw3 = np.matmul(db3.T, z2)                                        
    
    #Second Layer
    db2 = np.matmul(np.matmul(np.matmul(db3, W3).T, z2), (1-z2).T).T   
    dw2 = np.matmul(db2.T, z1) 

    #First Layer
    db1 = np.matmul(np.matmul(np.matmul(db2, W2).T, z1), (1-z1).T).T   
    dw1 = np.matmul(db1.T, x) 



    up_W5 = W5 - lr * dw5
    up_b5 = b5 - lr * db5
    up_W4 = W4 - lr * dw4
    up_b4 = b4 - lr * db4
    up_W3 = W3 - lr * dw3
    up_b3 = b3 - lr * db3
    up_W2 = W2 - lr * dw2
    up_b2 = b2 - lr * db2
    up_W1 = W1 - lr * dw1
    up_b1 = b1 - lr * db1
    

    return up_W1, up_b1, up_W2, up_b2, up_W3, up_b3, up_W4, up_b4, up_W5, up_b5
```

```python
import matplotlib.pyplot as plt


W1, b1, W2, b2, W3, b3, W4, b4, W5, b5 = initialize_weights_and_bias(mode='xavier', distribution='uniform')
target = np.array(train_labels)

cost = 0
losses = []
epochs = 50

for j in range(epochs):
    losss = 0
    tp = 0
    for i in range(len(train_image_list)):
        
        img = cv2.imread(train_image_list[i])
        x = preprocess_input(img)
        y = np.zeros(10)
        y[int(target[i])] = 1
        
    

        #network forward pass
        z1, z2, z3, z4, y_hat = forward_pass(x, W1, b1, W2, b2, W3, b3, W4, b4, W5, b5)
        
        
        #we use categorical cross entropy loss
        loss = 0
        for i in range(len(y)):
            if y[i] != 0:
                loss +=  -y[i] * np.log(y_hat[i])
        losss += loss
        
        
        #the true positives for this case, to calculate accuracy
        if np.argmax(y_hat) == np.argmax(y):
            tp += 1
        
        #network backward pass
        W1, b1, W2, b2, W3, b3, W4, b4, W5, b5 = backward_pass(x, y, W1, b1, W2, b2, W3, b3, W4, b4, W5, b5, z1, z2, z3, z4, y_hat)
    
    #we append total loss of each epoch to the losses array
    losses.append(losss)
    print(f"Epoch = {j}, Accuracy = {tp/len(train_image_list)}, loss= {losss}")
```

```python
# print(losses)
plt.plot(losses)
plt.show()
```
